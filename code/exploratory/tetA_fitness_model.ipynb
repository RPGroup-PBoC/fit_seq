## Exponential Growth Model

The objective of this notebook is to explore a biophysically inspired model for the fitness effect that tetracycline--a translation inhibitor--has on bacterial growth. The first model we will explore is where we assume that cells are happily growing exponentially at steady state (like on a chemostat for example). The number of cells $N$ changes over time as
$$
\frac{dN}{dt} = \lambda N.
$$
The growth rate $\lambda$ varies depending on the concentration of antibiotic in the media $a_{ex}$, as well as the intracellular drug concentration $a$. Given that tetracycline acts to inhibit translation by binding to the ribosomes, the variables we care about include the concentration of free ribosomes $r_u$ and of ribosomes bound by tetracycline $r_b$. The full system of ODE's defining this problem take the form
$$
\begin{array}{l}
\frac{\mathrm{d} a}{\mathrm{~d} t}=-\lambda a+f\left(r_{u}, r_{b}, a\right)+J\left(a_{\mathrm{ex}}, a\right), \\
\frac{\mathrm{d} r_{u}}{\mathrm{~d} t}=-\lambda r_{u}+f\left(r_{u}, r_{b}, a\right)+s(\lambda), \\
\frac{\mathrm{d} r_{b}}{\mathrm{~d} t}=-\lambda r_{b}-f\left(r_{u}, r_{b}, a\right),
\end{array}
$$

where $f(r_u, r_b, a)$ captures the binding-unbinding kinetics of the ribosome and the drug, and $J(a_{ex}, a)$ captures the import-export of the drug into and out of the cell. The ribosome-tetracycline binding kinetics are assumed to be a first-order reaction of the form
$$
f\left(r_{u}, r_{b}, a\right)=-k_{\mathrm{on}} a\left(r_{u}-r_{\mathrm{min}}\right)+k_{\mathrm{off}} r_{b},
$$
where $k_\mathrm{on}$ and $k_\mathrm{off}$ are the kinetic rates. For the import-export dynamics we also assume first-order reaction kinetics of the form
$$
J\left(a_{\mathrm{ex}}, a\right)=p_{\mathrm{in}} a_{\mathrm{ex}}-p_{\mathrm{out}} a,
$$
where $p_\mathrm{in}$ and $p_\mathrm{out}$ are the kinetic rates characterizing the import and export fluxes. As we will see later, we will assume that the efflux pump TetA actively increases $p_\mathrm{out}$.

Furthermore, to make progress in defining the growth rate as a function of the ribosome concentration we will make use of the empirical growth laws that Terrence Hwa's group has been advocating over the years. The laws are defined as
- **First Law**
$$
r_u = r_\mathrm{min} + \frac{\lambda}{\kappa_t}.
$$
- **Second Law**
$$
r_\mathrm{tot} = r_u + r_b = r_\max + \frac{\lambda}{\kappa_t} - \frac{\lambda \Delta r}{\lambda_o}.
$$

**NOTE: define variables**

After a lot of algebra (in my notes), we can arrive to a 3rd degree polynomial of the form
$$
\left(\frac{\lambda}{\lambda_{0}}\right)^{3}
-\left(\frac{\lambda}{\lambda_{0}}\right)^{2}
+\left(\frac{\lambda}{\lambda_{0}}\right)
\left[\frac{a_{e x}}{a^{*}}\left(\frac{\lambda_{0}^{*}}{\lambda_{0}}\right)
+\left(\frac{\lambda_{0}^{*}}{\lambda_{0}}\right)^{2}\right]
-\left(\frac{\lambda_{0}^{*}}{\lambda_{0}}\right)^{2}=0,
$$
where **NOTE: define variables**

Let us work with this 3rd degree polynomial with `sympy`. For this we need to define the variables

```python
# Define variables
lam = sympy.Symbol("\\left(\\frac{\lambda}{\lambda_0}\\right)")
lam_o = sympy.Symbol("\\left(\\frac{\lambda_0^*}{\lambda_0}\\right)")
a, aex = sympy.symbols("a^* a_{{\mathrm{ex}}}")

lam, lam_o, a, aex
```

Let us now define the polynomial

```python
# Define polynomial for lambda
lam_pol = sympy.Eq(
    lam ** 3 - lam ** 2 + lam * (aex / a * lam_o + lam_o ** 2) - lam_o ** 2, 0
)

lam_pol
```

We can find the roots of the polynomial by using `Sympy`'s solvers

```python
# Find roots of polynomial
lam_roots = sympy.solve(lam_pol, lam)
# lam_roots
```

These are obviously not trivial solutions, but we do not have to work with the messy algebra. We can convert these solutions into `lambdify` functions in order to evaluate them numerically.

```python
# Define variables in the right order to feed to function
var = [aex, a, lam_o]

# Generate list with lambdify functions
lam_roots_num = [sympy.lambdify(var, l, "scipy") for l in lam_roots]
```

With these functions we can explore numerically how the change in parameters affects the number of possible roots

```python
# Define sliders to change parameter values
lam_slider = pn.widgets.FloatSlider(name='/', start=0, end=5, step=0.0005, value=0.03)
a_slider = pn.widgets.FloatSlider(name='a', start=0.001, end=1, step=0.001, value=0.26)

@pn.depends(lam_slider.param.value, a_slider.param.value)
def lam_roots_plot(lam_slider, a_slider):
    """
    Function to generate interactive widget
    """
    # Define range of concentrations
    aex = np.linspace(0, 6, 1000)

    # Compute roots
    roots = [l(aex, a_slider, lam_slider) for l in lam_roots_num]
    
    # Initialize Bokeh figure
    p = bokeh.plotting.figure(
        height=400,
        width=500,
        x_axis_label="a (M)",
        y_axis_label="normalized growth rate ('/')"
    )
    # Define colors
    col = bokeh.palettes.Blues3

    # Add points
    for i, r in enumerate(roots):
        p.circle(
            x=aex[np.isclose(r.imag, 0)].real,
            y=r[np.isclose(r.imag, 0)].real,
            color=col[i],
        )
    # Set y axis limits
    p.y_range.start = -0.05
    p.y_range.end = 1.05
    # Apply PBoC style
    fit_seq.viz.pboc_single(p)
    
    return pn.pane.Bokeh(p)

pn.Column(
    lam_slider,
    a_slider,
    lam_roots_plot,
)
```

## Numerical approach

Let us now try an alternative strategy where, rather than analytically finding the roots of the 3rd degree polynomial, we use `numpy` to numerically find the values. For this we can use the `np.polynomial.Polynomial` function that very conveniently finds such roots. Let's define a function to find the roots.

```python
def lam_roots_np(aex, a, lam_o):
    '''
    Function to compute the roots of the 3rd degree
    polynomial.
    Parameters
    ----------
    aex : array-like.
        Antibiotic extenral concentrataions.
    a : float.
        Reference concentration (parameter from polynomial)
    lam_o : float.
        Reference growth-rate (parameter from polynomial)
    
    Returns
    -------
    roots : array-like. shape: len(aex) x 3
        Array with the polynomial roots
    '''
    # Initialize array to save roots
    # NOTE: Needs to allow for complex number entries
    roots = np.zeros([len(aex), 3], dtype=complex)
    
    # Loop through aex values
    for i, ae in enumerate(aex):
        # Define coefficients of polynomial of the form
        # A + Bx + Cx**2 + Dx**3 = 0
        A = -lam_o**2
        B = ae / a * lam_o + lam_o**2
        C = -1
        D = 1
        # Compute root
        roots[i, :] = np.polynomial.Polynomial([A, B, C, D]).roots()
        
    return roots
```

Let's repeat the interactive plot to make sure the numerics works as well.

```python
# Define sliders to change parameter values
lam_slider_np = pn.widgets.FloatSlider(name='/', start=0, end=50, step=0.0005, value=0.03)
a_slider_np = pn.widgets.FloatSlider(name='a', start=0.001, end=1, step=0.001, value=0.26)

@pn.depends(lam_slider_np.param.value, a_slider_np.param.value)
def lam_roots_plot_np(lam_slider_np, a_slider_np):
    """
    Function to generate interactive widget
    """
    # Define range of concentrations
    aex = np.linspace(0, 6, 1000)
    
    # Compute roots
    roots = lam_roots_np(aex, a_slider_np, lam_slider_np).T
    
    # Initialize Bokeh figure
    p = bokeh.plotting.figure(
        height=400,
        width=500,
        x_axis_label="a (M)",
        y_axis_label="normalized growth rate ('/')"
    )
    # Define colors
    col = bokeh.palettes.Blues3
    # Add points
    for i, r in enumerate(roots):
        p.circle(
            x=aex[np.isclose(r.imag, 0)].real,
            y=r[np.isclose(r.imag, 0)].real,
            color=col[i],
        )
    # Set y axis limits
    p.y_range.start = -0.05
    p.y_range.end = 1.05
    # Apply PBoC style
    fit_seq.viz.pboc_single(p)
    
    return pn.pane.Bokeh(p)

pn.Column(
    lam_slider_np,
    a_slider_np,
    lam_roots_plot_np,
)
```

# TetA resistance

Our experiments include the expression of a tetracycline-specific efflux pump *tetA*. The presence of this efflux pump and its non-linear Michaelis-Menten-like behavior changes the intracellular drug concentration. The resulting equations for the steady-state intracellular antibiotic concentration gives a quadratic equation of the form
$$
0=a^{2}\left[\frac{\lambda^{2}}{\kappa_{t} K_{D}}-\lambda-j\right]
+ a\left[\frac{K_{m} \lambda^{2}}{\kappa_{t} K_{D}}-K_{m} \lambda + j a_{e x}-j K_{m}-\frac{\lambda}{\lambda_{0}} V_{0}\right]
+ j a_{ex} K_m,
$$
where **DEFINE VARIABLES**. The roots of this quadratic equation are then plugged into the resulting equation that determines the growth rate
$$
0=-\frac{a^{*} \lambda}{\kappa_{t} K_{D}}+\Delta r\left(1-\frac{\lambda}{\lambda_{0}}\right),
$$
where **DEFINE VARIABLES**. 

Let us define these two equations in `sympy`. For this, we first define all of the variables.

```python
# Define variables
lam = sympy.Symbol("\\lambda")
lam_o = sympy.Symbol("\\lambda_o")
a, aex = sympy.symbols("a a_{{\mathrm{ex}}}")
kt, Kd, Km = sympy.symbols("\\kappa_t K_D K_M")
j, Vo, del_r = sympy.symbols("j V_o {{r}}")

lam, lam_o, a, aex, kt, Kd, Km, j, Vo, del_r
```

Let us now define both equations

```python
# Define equation 1
eq1 = (
    a ** 2 * (lam ** 2 / kt / Kd - lam - j)
    + a
    * (
        Km * lam ** 2 / kt / Kd
        - Km * lam
        + j * aex
        - j * Km
        - lam / lam_o * Vo
    )
    + j * aex * Km
)

# Compute second equation
eq2 = -a * lam / kt / Kd + del_r * (1 - lam / lam_o)

[sympy.Eq(eq1, 0), sympy.Eq(eq2, 0)]
```

We now solve the second equation for the drug concentration

```python
a_eq2 = sympy.solve(sympy.Eq(eq2, 0), a)[0]

a_eq2
```

We then substitute the solution into the first equation, and solve for the growth rate

```python
eq1_lam = sympy.Eq(eq1.subs(a, a_eq2), 0).expand()

eq1_lam
```

Multiply both sides by $\lambda^2$

```python
eq2_lam = sympy.Eq(eq1_lam.lhs * lam ** 2, 0).expand()

eq2_lam
```

Collect terms in powers of $\lambda$ to factorize a polynomial

```python
eq3_lam = sympy.Eq(
    sympy.collect(eq2_lam.lhs, [lam ** 4, lam ** 3, lam ** 2, lam, lam**0]), 0
)

eq3_lam
```

Let's now define a function that finds the roots of this 4th degree polynomial.

```python
def tetA_lam_roots(aex, lam_o, kappa_t, Kd, Km, j, Vo, del_r):
    """
    Function to numerically compute the roots of the 4th degree
    polynomial that defines the growth rate for the tetA system

    Parameters
    ----------
    aex : float.
        Extracellular antibiotic concentration
    lam_o : float.
        Growth rate in the absence of drug.
    kappa_t : float.
        Nutritional capacity.
    Kd : float.
        Drug-ribosome dissociation constant.
    Km : float.
        Drug-pump Michaelis-Menten constant.
    j : float.
        Membrane permeability to the drug.
    Vo : float.
        Expression of efflux pump under no antibiotic conditions.
    del_r : float.
        Dynamic range for the fraction of proteome made by ribosomes.
    """
    # Initialize array to save roots
    # NOTE: Needs to allow for complex number entries
    roots = np.zeros([len(aex), 4], dtype=complex)

    # Loop through aex values
    for i, ae in enumerate(aex):
        # Define elements of polynomial of the form
        # A + Bx + Cx**2 + Dx**3 + Ex**4 = 0
        A = -(Kd ** 2) * kappa_t ** 2 * j * del_r ** 2
        B = (
            -(Kd ** 2) * kappa_t ** 2 * del_r ** 2
            + 2 * Kd ** 2 * kappa_t ** 2 * j * del_r ** 2 / lam_o
            - Kd * Km * kappa_t * j * del_r
            + Kd * kappa_t * ae * j * del_r
        )
        C = (
            2 * Kd ** 2 * kappa_t ** 2 * del_r ** 2 / lam_o
            - Kd ** 2 * kappa_t ** 2 * j * del_r ** 2 / lam_o ** 2
            - Kd * Km * kappa_t * del_r
            + Kd * Km * kappa_t * j * del_r / lam_o
            - Kd * Vo * kappa_t * del_r / lam_o
            + Kd * kappa_t * del_r ** 2
            - Kd * kappa_t * ae * j * del_r / lam_o
            + Km * ae * j
        )
        D = (
            -(Kd ** 2) * kappa_t ** 2 * del_r ** 2 / lam_o ** 2
            + Kd * Km * kappa_t * del_r / lam_o
            + Kd * Vo * kappa_t * del_r / lam_o ** 2
            - 2 * Kd * kappa_t * del_r ** 2 / lam_o
            + Km * del_r
        )
        E = Kd * kappa_t * del_r ** 2 / lam_o ** 2 - Km * del_r / lam_o

        # Compute root
        roots[i, :] = np.polynomial.Polynomial([A, B, C, D, E]).roots()

    return roots
```

Let's make sure that the function works

```python
# Define parameters
aex = [1000]  # M
lam_o = 0.68  # h**-1
kappa_t = 0.06  # M**-1 h**-1
Kd = 1  # M
Km = 10  # M
j = 1  # h**-1
Vo = 0
del_r = 46.5  # M

# Pack parameters together
args = (aex, lam_o, kappa_t, Kd, Km, j, Vo, del_r)
# Find roots
tetA_lam_roots(*args)
```
